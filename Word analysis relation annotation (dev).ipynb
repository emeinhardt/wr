```python
def corpus_contexts_name(fp):
    if 'buckeye' in fp or 'Buckeye' in fp:
        return 'Buckeye'
    elif 'nxt_swbd' in fp or 'NXT_swbd' in fp:
        return 'NXT_swbd'
    else:
        raise Exception(f"corpus context in fp {fp} is neither buckeye nor nxt_swbd")
        
def get_contexts_direction(fp):
    if 'preceding' in fp:
        return 'preceding'
    elif 'following' in fp:
        return 'following'
    elif '1gram' in fp or '(empty)' in fp:
        return '(NA)'
    else:
        raise Exception(f"corpus context direction in fp {fp} is neither 'preceding' nor 'following' nor '(NA)' (= unigram)")
        
def get_contexts_order(fp):
#     direction = get_contexts_direction(fp)
    ngram_tokens = re.findall(r"[0-5]gram", fp)
    if len(ngram_tokens) < 1:
        raise Exception(f"No instance of substring matching '[0-5]gram' in {fp}; order could not be extracted.")
    orders = [token[0] for token in ngram_tokens]
    unique_orders = set(orders)
    if len(unique_orders) != 1:
        raise Exception(f"More than one 'n' for all substring tokens matching '[0-5]gram' in {fp}; no unique order could be extracted.")
    return list(unique_orders)[0]

def get_pseudocount(fp, abbrev=False):
    if abbrev:
        key_str = 'pc'
        pc_tokens = re.findall(r"pc0.[01]*", fp)
    else:
        key_str = 'pseudocount'
        pc_tokens = re.findall(r"pseudocount0.[01]*", fp)
#     pc_tokens = re.findall(r"pseudocount0.[01]*", fp)
    if len(pc_tokens) < 1:
        raise Exception(f"No instance of substring matching '{key_str}0.[01]*' in {fp}; pseudocount could not be extracted.")
    pcs = [token.split(key_str)[1] for token in pc_tokens]
    unique_pcs = set(pcs)
    if len(unique_pcs) != 1:
        raise Exception(f"More than one pseudocount for all substring tokens matching '{key_str}0.[01]*' in {fp}; no unique pseudocount could be extracted.")
    return list(unique_pcs)[0]

def get_lambda(fp, abbrev=False):
    if abbrev:
        key_str = 'l'
        l_tokens = re.findall(r"l[01][.][0-9]*", fp)
    else:
        key_str = 'lambda'
        l_tokens = re.findall(r"lambda[01][.][0-9]*", fp)
#     l_tokens = re.findall(r"lambda[01][.][0-9]*", fp)
    if len(l_tokens) < 1:
        raise Exception(f"No instance of substring matching '{key_str}[01][.][0-9]*' in {fp}; lambda could not be extracted.")
    ls = [token.split(key_str)[1] for token in l_tokens]
    unique_ls = set(ls)
    if len(unique_ls) != 1:
        raise Exception(f"More than one lambda for all substring tokens matching '{key_str}[01][.][0-9]*' in {fp}; no unique lambda could be extracted.")
    return list(unique_ls)[0]

# def get_pseudocount(fp):
#     pc_tokens = re.findall(r"pc0.[01]*", fp)
#     if len(pc_tokens) < 1:
#         raise Exception(f"No instance of substring matching 'pc0.[01]*' in {fp}; pseudocount could not be extracted.")
#     pcs = [token.split('pc')[1] for token in pc_tokens]
#     unique_pcs = set(pcs)
#     if len(unique_pcs) != 1:
#         raise Exception(f"More than one pseudocount for all substring tokens matching 'pc0.[01]*' in {fp}; no unique pseudocount could be extracted.")
#     return list(unique_pcs)[0]

# def get_lambda(fp):
#     l_tokens = re.findall(r"l[01][.][0-9]*", fp)
#     if len(l_tokens) < 1:
#         raise Exception(f"No instance of substring matching 'l[01][.][0-9]*' in {fp}; lambda could not be extracted.")
#     ls = [token.split('l')[1] for token in l_tokens]
#     unique_ls = set(ls)
#     if len(unique_ls) != 1:
#         raise Exception(f"More than one lambda for all substring tokens matching 'l[01][.][0-9]*' in {fp}; no unique lambda could be extracted.")
#     return list(unique_ls)[0]
```

```python
def get_and_validate_LTR_info(bundle):
    b = bundle
    pW_V_fp = get_dirs(b)['pW_V_fp']
    pW_V = importDict(pW_V_fp)
    Vs = set(pW_V.keys())
    Ws = set(join(set(walk_values(lambda pW_v: frozenset(pW_v.keys()),
                                  pW_V).values())))
    Ws_t = tuple(sorted(Ws))
    n_W = len(Ws_t)
    
    for each_v in pW_V:
        assert len(set(pW_V[each_v].keys())) == 1, f"Each orthographic word is assumed to map to a single unique segmental wordform: '{v}' maps to '[{set(pW_V[each_v].keys())}]'"
    
    V_to_W = {v:list(pW_V[v].keys())[0]
          for v in pW_V}
    
#     processed_orth_field = 'speech' if corpus_name == 'NXT_swbd' else 'orthographic_wordform'
    
#     segWord_field = 'segWord'
    
    LTR_info = {
        'pW_V_fp':pW_V_fp,
        'pW_V':pW_V,
        'Vs':Vs,
        'Ws':Ws,
        'Ws_t':Ws_t,
        'n_W':n_W,
        'V_to_W':V_to_W
    }
    return LTR_info
```

Every element (row, observation) of the word annotation relation either
 - does have a `segWord_field` and a `context_field`
 - doesn't have a `segWord_field` and a `context_field`

For any non-null pair of (`segWord`, `context`) values associated with at least one element of the word annotation relation, 
 - either both the `segWord` and `context` can be associated together by the model with a language model prior probability, or not
 - either both the `segWord` and `context` can be associated together by the model with a confusion model posterior probability, or not

Given a set of update bundles and a word analysis relation, then, we want to
 - associate every element of the relation with at most one segWord field
   - this means that all of the update bundles must involve lexicon transcription relations that cohere (i.e. no pair of update bundles have associated LTRs that assign the same orthographic word two distinct non-null segmental transcriptions)
 - extract just the set of all (context string, segWord) value pairs from the analysis relation

Then we can, for each update bundle in the set of consistent extant update bundles with posterior metadata,
 1. load the update bundle and its associated files
 2. map over the set of (context string, segWord) value pairs in the analysis relation and create an update object that assigns each such (context string, segWord) pair one (possibly null) value from the bundle's language model prior and another value for the bundle's confusion model posterior
   1. for this to be parallelizable, the "assignment" can't be stateful: it must be something representing a mapping from a (context string, segWord) pair to a (annotation field, probability/surprisal or null) pair.
   2. the parallel updates from different bundles (also language model vs. confusability assignments from the same bundle) should be easily mergeable: updates with the same (context string, segWord) should be trivially recognizable as capable of being merged.
   
Both `funcy` and `toolz` offer functions for merging dictionaries by combining key-value pairs with the same key (`merge` and `join`; `merge_with`).

Given a collection of $n$ sets of updates from $n$ update bundles, we can merge any pair of update collections to create a new (merged) update collection until we have a single collection of key-value mappings that map every (context, segWord) pair to a collection of (annotation field, annotation value = probability/surprisal or null) annotations, with at most one pair of language model + posterior annotations for each update bundle.

Given any update collection, we can then update the original word analysis relation by iterating through the word analysis relation, gathering all updates for each element of the analysis relation, and adding them to the analysis relation.

```python
# we need to use the loky backend rather than (the often way faster) multiprocessing backend
# because the multiprocessing backend uses the default pickle serialization module (vs. cloudpickle),
# and pickle doesn't support serialization of anonymous functions...
def lokypar(gen_expr):
    return Parallel(n_jobs=J, backend='loky', verbose=V, prefer=PREFER)(gen_expr)
```

```python
def my_parfold(g, z, xs):
    return lcompose(lokypar(delayed(curry(g))(x) for x in xs))(z)
```

```python
def jmap(f, xs):
    return par(delayed(f)(x) for x in xs)
```

```python
def create_annotation_update(Cs, Ws, Cs_t, Ws_t, pW_C, pW_WC, annotation_strings, input_triple):
    my_c_type = input_triple[0]
    my_c = input_triple[1]
    my_w = input_triple[2]

    lm_str_p = annotation_strings['lm_str_p']
    lm_str_h = annotation_strings['lm_str_h']
#     lm_str_h_bar = annotation_strings['lm_str_h_bar']
    post_str_p = annotation_strings['post_str_p']
    post_str_h = annotation_strings['post_str_h']
#     post_str_h_bar = annotation_strings['post_str_h_bar']
    
    modelable_c = my_c in Cs
    modelable_w = my_w in Ws
    
    result = []
    if not modelable_c:
        result = ['bad_c']
    if not modelable_w:
        result += ['bad_w']
    
    update_rel = dict()
    
    if len(result) > 0:
        result = str_join(',', result)
        update_rel[lm_str_p] = result
        update_rel[lm_str_h] = result
#         update_rel[lm_str_h_bar] = result
        update_rel[post_str_p] = result
        update_rel[post_str_h] = result
#         update_rel[post_str_h_bar] = result
        update = {input_triple:update_rel}
        return update
    
    c_idx = idx(my_c, Cs_t)
    w_idx = idx(my_w, Ws_t)
    
    lm_result = pW_C[w_idx, c_idx]
    post_result = pW_WC[w_idx, c_idx]
    
    if lm_result == np.inf:
        lm_result = float(np.nan)
        update_rel[lm_str_p] = lm_result
        update_rel[lm_str_h] = lm_result
    elif lm_result < 0.0:
        raise Exception(f"lm result < 0!\nctxt:{my_c}\nmy_w:{my_w}\nlm result:{lm_result}")
    elif lm_result > 1.0 and not np.allclose(np.array([lm_result - 1]), np.zeros(np.array([lm_result - 1]).shape), atol=1e-06):
#         lm_deviation = np.array([lm_result - 1])
#         np.allclose(lm_deviation, np.zeros(lm_deviation.shape), atol=1e-06)
        raise Exception(f"lm result > 1!\nctxt:{my_c}\nmy_w:{my_w}\nlm result:{lm_result}")
    else:
        update_rel[lm_str_p] = lm_result
        update_rel[lm_str_h] = -1.0 * float(np.log2(lm_result))
    
    if post_result == np.inf:
        post_result = float(np.nan)
        update_rel[post_str_p] = post_result
        update_rel[post_str_h] = post_result
    elif post_result < 0.0:
        raise Exception(f"post result < 0!\nctxt:{my_c}\nmy_w:{my_w}\nlm result:{lm_result}\npost_result{post_result}")
    elif post_result > 1.0 and not np.allclose(np.array([post_result - 1]), np.zeros(np.array([post_result - 1]).shape), atol=1e-06):
#         post_deviation = np.array([post_result - 1])
#         np.allclose(post_deviation, np.zeros(post_deviation.shape), atol=1e-06)
        raise Exception(f"post result > 1!\nctxt:{my_c}\nmy_w:{my_w}\nlm result:{lm_result}\npost_result{post_result}")
    else:
        update_rel[post_str_p] = float(post_result)
        update_rel[post_str_h] = -1.0 * float(np.log2(post_result))
    
    update = {input_triple:update_rel}
    return update
```

```python
def annotation_lookup(context, segmental_wordform, update_bundle, LTR_info=None):
    b = update_bundle
    my_c = context
    my_w = segmental_wordform
    
    dir_info = get_dirs(b)
    post_fp = dir_info['pW_WC_dir']
    print(f'post_fp = {post_fp}')
    
#     startNote(f'collecting information for updates from {post_fp}')
    
    context_info = get_context_info(b)
#     print(f"context description:\n{get_context_description(b)}")
    context_description = get_context_description(b)
    context_field = context_description['context_field']
    validate_context_info(b, context_info)
    
    if LTR_info is None:
        LTR_info = retrieve_precomputed_LTR_info(b)
    
    LM_info = get_LM_info(b)
    
    posterior_info = get_posterior_info(b)
    posterior_validation_info = validate_posterior(posterior_info, LTR_info, context_info)
    
    annotation_strings = get_annotation_strings(b)
    print(f"Annotation strings = \n{annotation_strings}")
    
#     Cs, Ws, Cs_t, Ws_t, pW_C, pW_WC, annotation_strings
    Cs = context_info['Cs']
    Ws = LTR_info['Ws']
    Cs_t = context_info['Cs_t']
    Ws_t = LTR_info['Ws_t']
    pW_C = LM_info['pW_C']
    pW_WC = posterior_info['pW_WC']
    
    print(f'|Cs| = {len(Cs)}')
    my_c_idx = Cs_t.index(my_c)
    my_w_idx = Ws_t.index(my_w)
    
    pw_wc = pW_WC[my_w_idx, my_c_idx]
    
    return pw_wc

def make_annotation_lookup_dict_for(update_bundle, LTR_info=None):
    b = update_bundle
    
    dir_info = get_dirs(b)
    post_fp = dir_info['pW_WC_dir']
    print(f'post_fp = {post_fp}')
    
#     startNote(f'collecting information for updates from {post_fp}')
    
    context_info = get_context_info(b)
#     print(f"context description:\n{get_context_description(b)}")
    context_description = get_context_description(b)
    context_field = context_description['context_field']
    validate_context_info(b, context_info)
    
    if LTR_info is None:
        LTR_info = retrieve_precomputed_LTR_info(b)
    
    LM_info = get_LM_info(b)
    
    posterior_info = get_posterior_info(b)
    posterior_validation_info = validate_posterior(posterior_info, LTR_info, context_info)
    
    annotation_strings = get_annotation_strings(b)
    print(f"Annotation strings = \n{annotation_strings}")
    
#     Cs, Ws, Cs_t, Ws_t, pW_C, pW_WC, annotation_strings
    Cs = context_info['Cs']
    Ws = LTR_info['Ws']
    Cs_t = context_info['Cs_t']
    Ws_t = LTR_info['Ws_t']
    pW_C = LM_info['pW_C']
    pW_WC = posterior_info['pW_WC']
    
    def lookup(context, segmental_wordform):
        my_c = context
        my_w = segmental_wordform
        
        if my_c not in Cs_t:
            print(f'bad c: {my_c}')
        if my_w not in Ws_t:
            print(f'bad w: {my_w}')
        my_c_idx = Cs_t.index(my_c)
        my_w_idx = Ws_t.index(my_w)

        pw_wc = pW_WC[my_w_idx, my_c_idx]
        
        return pw_wc
    
    return {'lookup':lookup, 
            'Cs_t':Cs_t,
            'Ws_t':Ws_t,
            'pW_C':pW_C,
            'pW_WC':pW_WC,
            'context_field':context_field}

def test_update_collection(update_collection, update_bundle, tests=1000):
    my_lookup_bundle = make_annotation_lookup_dict_for(update_bundle)
    pW_V_fp = get_dirs(update_bundle)['pW_V_fp']
    
    lookup = my_lookup_bundle['lookup']
    
    Cs_t = my_lookup_bundle['Cs_t']
    Ws_t = my_lookup_bundle['Ws_t']
    
    context_description = get_context_description(update_bundle)
    context_field = context_description['context_field']
    annotation_strings = get_annotation_strings(update_bundle)
#     lm_annotation_string
    post_annotation_string = annotation_strings['post_str_h']
    
    keys_to_check = choices(lfilter(lambda triple: first(triple) == context_field,
                                    update_collection.keys()), 
                            k=tests)
    error_count = 0
    for k in tqdm(keys_to_check):
        c_to_check, w_to_check = second(k), third(k)
        result_to_check = update_collection[k][post_annotation_string]
        correct_result = lookup(c_to_check, w_to_check)
        
        
        if correct_result != result_to_check:
            error_count += 1
            print(f'pW_V_fp:\n\t{pW_V_fp}')
            print(f"context field:\n\t{context_field}")
            print(f"annotation string:\n\t{post_annotation_string}")
            print(f"Key:\n\t{k}")
            print(f"Whole result for key:\n\t{update_collection[k]}")
            print(f"Result in update collection:\n\t{result_to_check}")
            print(f"Correct result:\n\t{correct_result}")
            print('----------------------------------------')
    print(f'\nError count: {error_count}\n')
    
```

```python
segWord_to_annotation_to_surprisals = dict()

def associate_surprisal_measures(context_segword_triple):
    #STATEFUL UPDATE
    my_w = context_segword_triple[2]    
    my_updates = merged_update_collections[context_segword_triple]
    
    if my_w not in segWord_to_annotation_to_surprisals:
        segWord_to_annotation_to_surprisals[my_w] = dict()
    
    existing_measures = segWord_to_annotation_to_surprisals[my_w]
    
    for annotation_field in my_updates:
        my_annotation_value = deepcopy(my_updates[annotation_field])
        
        if annotation_field not in existing_measures and type(my_annotation_value) != type("foo"):
            segWord_to_annotation_to_surprisals[my_w][annotation_field] = {my_annotation_value}
        elif annotation_field in existing_measures and type(my_annotation_value) != type("foo"):
            existing_vals = segWord_to_annotation_to_surprisals[my_w][annotation_field]
            segWord_to_annotation_to_surprisals[my_w][annotation_field] = merge(existing_vals, {my_annotation_value})

for context_segword_triple in tqdm(merged_update_collections):
    associate_surprisal_measures(context_segword_triple)
```

```python
my_segWords_for_avgs = set(segWord_to_annotation_to_surprisals.keys())

segWord_to_annotation_to_avg_surprisal = dict()

def calc_avg_measure(vals):
    n_contexts_with_w_and_measure = len(vals)
    avg_val = sum(vals) / n_contexts_with_w_and_measure if n_contexts_with_w_and_measure != 0 else 'bad_w'
    return avg_val    

def to_avg_field(annotation_field):
    af = annotation_field
    if '-bar' in af:
        return af
    if af[:2] == 'p(':
        return 'p-bar' + af[1:]
    if af[:2] == 'h(':
        return 'h-bar' + af[1:]
    raise Exception(f"field = {af}")

def avg_prob_measure_calculation(segWord):
    #STATEFUL UPDATE
    my_w = segWord
    if my_w in segWord_to_annotation_to_avg_surprisal:
        raise Exception('This function should only be called once for each segWord.')
    
    my_measures = deepcopy(segWord_to_annotation_to_surprisals[my_w])
    my_renamed_measures = walk_keys(to_avg_field, my_measures)
    my_avgs = walk_values(calc_avg_measure, my_renamed_measures)
    
    segWord_to_annotation_to_avg_surprisal[my_w] = my_avgs
    
#     return my_avgs
    
    
#     my_surprisals = annotation_to_segWord_to_surprisals

for each_w in tqdm(my_segWords_for_avgs):
    avg_prob_measure_calculation(each_w)
```

```python
def neighborhood_update(rel):
    new_rel = deepcopy(rel)
    my_w = new_rel[segWord_field]
    
    if my_w is None:
        return new_rel
    
    if my_w in neighborhood_size_dict:
        new_rel['Neighborhood size'] = neighborhood_size_dict[my_w]
    else:
        new_rel['Neighborhood size'] = np.nan
    
    if my_w in neighborhood_density_dict:
        new_rel['Weighted neighborhood density'] = neighborhood_density_dict[my_w]
    else:
        new_rel['Weighted neighborhood density'] = np.nan
    
    return new_rel

neighborhood_updated_WAR = par(delayed(neighborhood_update)(rel)
                               for rel in fully_updated_word_analysis_relation)
```

# Add phonotactic surprisal information

```python
nphone_tsv_fp = 'LTR_Buckeye_aligned_w_GD_AmE_destressed/LTR_Buckeye_aligned_CM_filtered_word_type_pW_nphone.tsv'
```

```python
pW_nphone_relation = loadTSV_as_dictlist(nphone_tsv_fp, ['w', 'surprisal'])
pW_nphone_relation[:5]
pW_nphone_relation[1:5]
pW_nphone_relation = pW_nphone_relation[1:]
```

```python
def phonotactic_prob_update(rel):
    if segWord_field not in rel:
        return rel
    my_w = rel[segWord_field]
    matches = lfilter(lambda row: row['w'] == my_w,
                      pW_nphone_relation)
    if len(matches) == 0:
        nphone_surp = np.nan
    elif len(matches) > 1:
        raise Exception(f"{my_w}, {len(matches)}\n\t{matches}\n\t{rel}")
    else:
        match = matches[0]
        nphone_surp = match['surprisal']
    rel['smoothed_5phone_prob'] = nphone_surp
    return rel
```

```python
phonotactic_prob_updated_WAR = par(delayed(phonotactic_prob_update)(rel)
                                   for rel in neighborhood_updated_WAR)
```
