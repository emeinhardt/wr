## Identify contexts of each symbol

```python
h_triph_t = list(map(dottedStringToTuple, h_triph))
i_triph_t = list(map(dottedStringToTuple, i_triph))
h_triph_t[:10]
i_triph_t[:10]
```

```python
def getContexts(symbol, tuples):
    contexts = set([t for t in tuples if t[1] == symbol])
    return contexts

def getHoleContexts(symbol, tuples):
    hole_contexts = set([(t[0], t[2]) for t in tuples if t[1] == symbol])
    return hole_contexts

def getLeftContexts(symbol, tuples):
    return set([t[0] for t in tuples if t[1] == symbol])

def getRightContexts(symbol, tuples):
    return set([t[2] for t in tuples if t[1] == symbol])

h_context_map = {s:getHoleContexts(s, h_triph_t) for s in h_symbols}
i_context_map = {s:getHoleContexts(s, i_triph_t) for s in i_symbols}
```

## Center and Context Distributions

```python
import math

def log(x):
    if x == 0.0:
        return 0
    return math.log2(x)

def h(x):
    return -1.0 * log(x)

def H(dist):
    return sum(dist[each]*h(dist[each]) for each in dist)
```

Are the response diphones a subset of the stimuli diphones?

```python
hammond_destressed_diphones_stim_fn = 'Hammond-aligned_destressed stimuli diphones.txt'
hammond_destressed_diphones_resp_fn = 'Hammond-aligned_destressed response diphones.txt'
iphod_destressed_diphones_stim_fn = 'IPhOD-aligned_destressed stimuli diphones.txt'
iphod_destressed_diphones_resp_fn = 'IPhOD-aligned_destressed response diphones.txt'
```

```python
h_stim_diph = []
with open(hammond_destressed_diphones_stim_fn, 'r') as the_file:
    for row in the_file:
        h_stim_diph.append(row.rstrip('\r\n'))
h_resp_diph = []
with open(hammond_destressed_diphones_stim_fn, 'r') as the_file:
    for row in the_file:
        h_resp_diph.append(row.rstrip('\r\n'))

i_stim_diph = []
with open(iphod_destressed_diphones_stim_fn, 'r') as the_file:
    for row in the_file:
        i_stim_diph.append(row.rstrip('\r\n'))
i_resp_diph = []
with open(iphod_destressed_diphones_resp_fn, 'r') as the_file:
    for row in the_file:
        i_resp_diph.append(row.rstrip('\r\n'))
```

```python
i_diph_pair = {'stim':set(i_stim_diph),'resp':set(i_resp_diph)}
h_diph_pair = {'stim':set(h_stim_diph),'resp':set(h_resp_diph)}
diph_pairs = [i_diph_pair, h_diph_pair]
```

```python
def pairReport(pair):
    print('# stimuli diphones: {0}'.format(len(pair['stim'])))
    print('# response diphones: {0}'.format(len(pair['resp'])))
    print('# resp diphones not in stimuli = {0}'.format(len(pair['resp'] - pair['stim'])))
    print('# stim diphones not in responses = {0}'.format(len(pair['stim'] - pair['resp'])))

pairReport(i_diph_pair)
print('----')
pairReport(h_diph_pair)
```
